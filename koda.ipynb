```python
def NarisiGraf(graf):
    razdalje = graf.distance_all_pairs()
    l = razdalje.values()
    k = []
    for i in l:
        for j in i.values():
            k = k + [j]
    pojavitve = {}
    for i in k:
        if i in pojavitve:
            pojavitve[i] += 1
        else:
            pojavitve[i] = 1
    lists = sorted(pojavitve.items())
    x, y = zip(*lists)
    r = list(y)
    for i in range(len(r)):
            r[i] = r[i]/2
    y = tuple(r)
    y = y[1:len(y)]
    x = x[1:len(x)]
    plt.plot(x, y)
```

```python
def PorazdelitevRazdalj(grafi):
    podatki = []
    for graf in grafi:
        NarisiGraf(graf)
```
